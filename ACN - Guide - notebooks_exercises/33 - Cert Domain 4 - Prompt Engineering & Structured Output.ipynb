##### Prompt (Severity Definitions)
"Severity definitions:

  in production under normal usage. Example: writing user password
  to application log in plaintext.

HIGH: Will cause incorrect behavior that users will notice.
  Example: off-by-one error in pagination causing items to be

MEDIUM: Could cause issues under specific conditions that are

  This does NOT meet criteria for EMERGENT because ECG changes are
  nonspecific, not diagnostic.
Classification: URGENT

  but nonspecific ECG prevents EMERGENT classification.
