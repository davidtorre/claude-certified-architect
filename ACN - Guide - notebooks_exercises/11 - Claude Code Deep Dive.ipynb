# .claude/commands/deploy.md
    # Deploy to Staging

    Run the following deployment steps in order:

    1. Run all tests: `npm test`
    2. Build the project: `npm run build`
    3. Run the staging deploy script: `./scripts/deploy-staging.sh`
    4. Verify the deployment by checking the health endpoint
    5. Report the results

    If any step fails, stop and report the error. Do not continue.

    name: Code Review Standards
    description: Enforces team code review standards and checklist
    tags: [review, quality, standards]
    operations: [review, audit, check]
    context: fork
    allowed-tools: [Read, Grep, Glob]
    argument-hint: File path or PR number to review
    ---

    # Code Review Checklist

    When reviewing code, check the following:

    ## Security
    - No hardcoded secrets, API keys, or credentials
    - Input validation on all user-facing endpoints
    - SQL queries use parameterized statements

    ## Code Quality
    - Functions are under 50 lines
    - No duplicate code blocks (DRY)
    - Clear, descriptive variable and function names
    - Complex logic has explanatory comments

    ## Testing
    - New code has corresponding test coverage
    - Edge cases are tested
    - Tests follow the Arrange-Act-Assert pattern

    ## Performance
    - No N+1 database queries
    - Large lists use pagination
    - Expensive operations are cached where appropriate

# Simple one-shot usage
    claude -p "Explain what this project does in 3 sentences"
# Pipe input to Claude
    git diff | claude -p "Review this diff for bugs"
# JSON output
    claude -p --output-format json \
    "List all API endpoints with their methods"
# Enforce schema

# .github/workflows/claude-review.yml
    name: Claude Code Review
    on:
    pull_request:
        types: [opened, synchronize]

    jobs:
    review:
        runs-on: ubuntu-latest
        steps:
        - uses: actions/checkout@v4
            with:
            fetch-depth: 0

        - name: Claude Code Review
            run: |
            git diff origin/main...HEAD | \
            claude -p --output-format json \
                "Review this diff. Report bugs, security issues,
                and suggestions. Be specific about file and line."
            env:
            ANTHROPIC_API_KEY: ${{ secrets.ANTHROPIC_API_KEY }}
